```python
books["Id"] = books["Id"].astype(int)
```

```python
books["Customer ID"] = books["Customer ID"].astype(int)
```

```python
books.columns = books.columns.str.strip()
```

```python
#remove "" from Book checkout
books["Book checkout"]=books["Book checkout"].astype(str).str.replace('"', '', regex=False).str.strip()
```

```python
#convert days allowed to borrow from 2 weeks to days
books["Days allowed to borrow"] = (books["Days allowed to borrow"].astype(str)
.str.extract(r"(\d+)", expand=False)
.astype(int) * 7)
```

```python
#convert book checkout data to date time 
books["Book checkout"] = pd.to_datetime(
    books["Book checkout"],
    dayfirst=True,
    errors="coerce"
)
```

```python
books["Book Returned"] = pd.to_datetime(
    books["Book Returned"],
    dayfirst=True,
    errors="coerce"
)
```
